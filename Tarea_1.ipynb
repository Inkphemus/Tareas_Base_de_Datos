## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Director"

```python
%%sql
CREATE TABLE director (id_director INT GENERATED ALWAYS AS IDENTITY,
nombre VARCHAR(45),
edad INT,
PRIMARY KEY(id_director))
```

### Tabla "Actor"

```python
%%sql
CREATE TABLE actor (id_actor INT GENERATED ALWAYS AS IDENTITY,
nombre VARCHAR(45),
edad INT,
PRIMARY KEY(id_actor))
```

### Tabla "Pelicula"

```python
%%sql
CREATE TABLE pelicula (id_pelicula INT GENERATED ALWAYS AS IDENTITY,
id_director INT,
nombre VARCHAR(45),
genero VARCHAR(45),
duracion INT,
fecha_publicacion DATE,
PRIMARY KEY (id_pelicula),
FOREIGN KEY (id_director) REFERENCES director(id_director))
```

### Tabla "actor_participa_en_pelicula"

```python
%%sql
CREATE TABLE actor_participa_en_pelicula (id_actor INT,
                                          id_pelicula INT,
                                          PRIMARY KEY (id_actor, id_pelicula),
                                          FOREIGN KEY (id_actor) REFERENCES actor(id_actor),
                                          FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula))
```

### Tabla  "Premio_mejor_director"

```python
%%sql
CREATE TABLE premio_mejor_director (id_premio INT GENERATED ALWAYS AS IDENTITY,
id_director INT,
anio INT,
PRIMARY KEY (id_premio), FOREIGN KEY(id_director) REFERENCES director(id_director))
```

### Tabla  "Premio_mejor_actor"

```python
%%sql
CREATE TABLE premio_mejor_actor (id_premio INT GENERATED ALWAYS AS IDENTITY,
id_actor INT,
anio INT,
PRIMARY KEY (id_premio),
FOREIGN KEY(id_actor) REFERENCES actor(id_actor))
```

### Tabla  "Premio_mejor_pelicula"

```python
%%sql
CREATE TABLE premio_mejor_pelicula (id_premio INT GENERATED ALWAYS AS IDENTITY,
id_pelicula INT,
anio INT,
PRIMARY KEY (id_premio), FOREIGN KEY(id_pelicula) REFERENCES pelicula(id_pelicula))
```

### Consulta 6
**Nombre de la película con la mayor brecha de edad entre el actor más joven y el actor más viejo que participen en ella, y las edades de dichos actores.** (Si hay empate, mostrar solamente uno).

**Solución:** _Escriba aquí la descripción de su solución._

```python
%%sql
SELECT t_p.nombre AS nombre_pelicula, 
           MAX(t_a.edad) AS edad_actor_mayor,
           MIN(t_a.edad) AS edad_actor_menor
           
  FROM pelicula t_p
    
  INNER JOIN actor_participa_en_pelicula t_ap ON
             t_p.id_pelicula = t_ap.id_pelicula
  INNER JOIN actor t_a ON 
             t_ap.id_actor = t_a.id_actor
             
    GROUP BY t_p.nombre
    ORDER BY ABS(MAX(t_a.edad) - MIN(t_a.edad))
    DESC
    LIMIT 1
```

**Solución:** _Escriba aquí la descripción de su solución._

```python
%%sql
WITH
tabla1 AS(
    SELECT t_a.nombre, t_a.id_actor, COUNT(DISTINCT t_p.id_director) AS dir_num
    FROM director t_d
    INNER JOIN pelicula t_p ON
        t_d.id_director = t_p.id_director
    INNER JOIN actor_participa_en_pelicula t_ap ON
        t_p.id_pelicula = t_ap.id_pelicula
    INNER JOIN actor t_a ON 
        t_ap.id_actor = t_a.id_actor
    GROUP BY t_a.id_actor
    HAVING COUNT(DISTINCT t_p.id_director) > 1
),
tabla2 AS(
    SELECT pp.nombre AS nombre_de_una_pelicula, tabla1.id_actor
    FROM tabla1
    INNER JOIN actor_participa_en_pelicula app ON
        tabla1.id_actor = app.id_actor
    INNER JOIN pelicula pp ON
        app.id_pelicula = pp.id_pelicula
),
tabla3 AS(
    SELECT tabla1.nombre, tabla1.dir_num, tabla2.nombre_de_una_pelicula
    FROM tabla1
    INNER JOIN tabla2 ON
        tabla1.id_actor = tabla2.id_actor
)
SELECT * FROM tabla3
```

### Consulta 9

**Nombre de todos los actores que han participado sólo en películas dirigidas por el director con id\_director 1.**


**Ejemplo:**
|Nombre      |
|------------|
|Ryan Gosling|
|...         |


**Solución:** _Escriba aquí la descripción de su solución._

```python
%%sql
SELECT t_a.nombre

  FROM director t_d

  INNER JOIN pelicula t_p ON
             t_d.id_director = t_p.id_director

  INNER JOIN actor_participa_en_pelicula t_ap ON
             t_p.id_pelicula = t_ap.id_pelicula

  INNER JOIN actor t_a ON 
             t_ap.id_actor = t_a.id_actor

  LEFT JOIN 
    (
        SELECT 
            id_actor,
            id_director
        FROM 
            actor_participa_en_pelicula ap
        INNER JOIN 
            pelicula p ON ap.id_pelicula = p.id_pelicula
        WHERE 
            id_director != 1
    ) out ON t_ap.id_actor = out.id_actor

  WHERE 
    t_d.id_director = 1 AND
    out.id_director IS NULL
  GROUP BY 
    t_a.nombre
```

```python
%%sql
WITH
--Crear una tabla con los nombres e id de los actores sin premios
tabla1 AS(
    SELECT nombre, actor.id_actor
    FROM actor
    FULL JOIN premio_mejor_actor p_ma ON
       actor.id_actor = p_ma.id_actor
    WHERE p_ma IS NULL
),
--Juntamos la tabla anterior con actor_participa_en_pelicula
tabla2 AS(
    SELECT tabla1.nombre, a_pp.id_pelicula
    FROM tabla1
    INNER JOIN actor_participa_en_pelicula a_pp ON
        tabla1.id_actor = a_pp.id_actor
),
--Juntamos la tabla anterior con las peliculas para obtener sus nombres
tabla3 AS(
    SELECT tabla2.nombre, pp.nombre AS nombre_pelicula, pp.id_director
    FROM tabla2
    INNER JOIN pelicula pp ON
        tabla2.id_pelicula = pp.id_pelicula
),
--Juntamos la tabla anterior con los directores de dichas peliculas para obtener sus nombres
tabla4 AS(
    SELECT tabla3.nombre, tabla3.nombre_pelicula, tabla3.id_director, dir.nombre AS nombre_director
    FROM tabla3
    INNER JOIN director dir ON
        tabla3.id_director = dir.id_director
),
--Tabla auxiliar para obtener el número de peliculas de un director
tabla5 AS(
    SELECT DISTINCT tabla4.id_director, COUNT(DISTINCT pp.id_pelicula) AS peliculas_del_director
    FROM tabla4
    INNER JOIN pelicula pp ON
        pp.id_director = tabla4.id_director
	GROUP BY tabla4.id_director
),
--Tabla auxiliar para obtener el número de premios de un director
tabla6 AS(
    SELECT DISTINCT tabla4.id_director, COUNT(DISTINCT pmd.anio) AS premios_del_director
    FROM tabla4
    INNER JOIN premio_mejor_director pmd ON
        pmd.id_director = tabla4.id_director
	GROUP BY tabla4.id_director
),
--Juntamos la tabla 4 con la cantidad de premios de cada director
tabla7 AS(
    SELECT tabla4.nombre, tabla4.nombre_pelicula, tabla4.nombre_director, tabla4.id_director, COALESCE(tabla6.premios_del_director, 0) AS numero_de_premios_del_director
    FROM tabla4
	LEFT JOIN tabla6 ON
		tabla4.id_director = tabla6.id_director
),
--Juntamos la tabla anterior con la cantidad de peliculas de cada director
tabla8 AS(
    SELECT tabla7.nombre, tabla7.nombre_pelicula, tabla7.nombre_director, COALESCE(tabla5.peliculas_del_director, 0) AS numero_de_peliculas_del_director, tabla7.numero_de_premios_del_director
    FROM tabla7
	LEFT JOIN tabla5 ON
		tabla5.id_director = tabla7.id_director
)
--Print
SELECT * from tabla8
```
